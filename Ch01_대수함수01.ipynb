##Sigmoid Function

image.png

```python
from IPython.display import display, Latex

display(Latex(r'$\displaystyle y = \frac{1}{1 + e^{-x}}$'))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
  return (1 / (1 + np.exp(-x)))

x = np.arange(-10,10,0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Sigmoid Function')
# plt.grid(True)
plt.show()
```

```python
from sympy import symbols, integrate, sin, pi
from IPython.display import display, Latex
import matplotlib.pyplot as plt
import numpy as np

# 수식 출력
display(Latex(r'$\displaystyle \int_{0}^{\pi/3} \sin(x)\, dx$'))

# 그래프 그리기
x = np.linspace(0, np.pi, 1000)
y = np.sin(x)
plt.plot(x, y, label='sin(x)', color='blue')
plt.fill_between(x[x <= np.pi/3], y[x <= np.pi/3], alpha=0.3, label=r'$\int_0^{\pi/3} \sin(x)\, dx$')
plt.legend()
plt.grid(True)
plt.title("Visualization of Integral")
plt.show()
```

## base2

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 범위 설정
x = np.linspace(-10, 10, 400)
y = 2 * x**2 + 3 * x - 1

# 그래프 그리기
plt.plot(x, y, label=r'$y = 2x^2 + 3x - 1$', color='green')
plt.title("Quadratic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

## Section 02 일차함수와 이차함수


### 일차함수
y= ax+b

a : slope (기울기)

b : y-intercept (y절편)

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Latex

# 1. 기울기 a, y절편 b 설정
a = 2
b = 1

# 2. 수식 출력 (LaTeX)
display(Latex(fr"$y = {a}x + {b}$"))

# 3. x 값 생성
x = np.linspace(-10, 10, 400)

# 4. y 값 계산
y = a * x + b

# 5. 그래프 그리기
plt.plot(x, y, label=fr'$y = {a}x + {b}$', color='blue')
plt.title("Linear Function: y = ax + b")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

###이차함수

```python
from IPython.display import display, Latex
display(Latex(r"$y = x^2$"))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Latex

# 수식 LaTeX로 출력
display(Latex(r"$y = x^2$"))

# x 값 범위 설정
x = np.linspace(-10, 10, 400)

# y = x^2 함수 정의
y = x ** 2

# 그래프 그리기
plt.plot(x, y, label=r'$y = x^2$', color='green')
plt.title("Quadratic Function: y = x²")
plt.xlabel("x")
plt.ylabel("y")
# plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
