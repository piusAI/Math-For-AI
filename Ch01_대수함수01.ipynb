## Section 02 일차함수와 이차함수


### 일차함수
y= ax+b

a : slope (기울기)

b : y-intercept (y절편)

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Latex

# 1. 기울기 a, y절편 b 설정
a = 2
b = 1

# 2. 수식 출력 (LaTeX)
display(Latex(fr"$y = {a}x + {b}$"))

# 3. x 값 생성
x = np.linspace(-10, 10, 400)

# 4. y 값 계산
y = a * x + b

# 5. 그래프 그리기
plt.plot(x, y, label=fr'$y = {a}x + {b}$', color='blue')
plt.title("Linear Function: y = ax + b")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

###이차함수

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Latex

# 수식 LaTeX로 출력
display(Latex(r"$y = x^2$"))

# x 값 범위 설정
x = np.linspace(-10, 10, 400)

# y = x^2 함수 정의
y = x ** 2

# 그래프 그리기
plt.plot(x, y, label=r'$y = x^2$', color='green')
plt.title("Quadratic Function: y = x²")
plt.xlabel("x")
plt.ylabel("y")
# plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
